## Word Tokenization

```python
tokenized_str = string.split()
```

```python
print(tokenized_str)
```

```python
token_word2idx = {ch:idx for idx, ch in enumerate(sorted(set(tokenized_str)))}
```

```python
print(token_word2idx)
```

```python
input_ids = [token_word2idx[token] for token in tokenized_str]
```

```python
print(input_ids)
```

# Subword Tokenization


1. WordPiece - 使用基本的 Unicode 字符來建立 word 的库存, 后构建语言模型，之后把库存里面的两个word 合并生成新的word放回， 一直循环直到符合既定阈值


2. Byte Pair Encoding -  GPT 家族使用的算法。通过一个基础的word list， 对每个word 进行字符级别的排列组合，生成新的word


3. Unigram - 从一个巨大的词汇表开始，逐个删减来收缩。

## Hugging face tokenizer

```python
from transformers import AutoTokenizer
```

```python
string = "Only those who will risk going too far can possibly find out how far one can go."
```

```python
model_name = 'distilbert-base-uncased-finetuned-sst-2-english'
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
encoded_str = tokenizer(string, padding=True, truncation=True)
```

```python
encoded_str
```

Special Token ID 下表：

Special Token	[PAD]	[UNK]	[CLS]	[SEP]	[MASK]

Special Token ID	0	100	101	102	103

```python
tokens = tokenizer.convert_ids_to_tokens(encoded_str.input_ids)
```

```python
tokens
```

```python
type(encoded_str)
```

```python
print(tokenizer.convert_tokens_to_string(tokens))
```

```python
from datasets import load_dataset
```

```python
sentiment = load_dataset("poem_sentiment")
```

```python
def tokenize(batch):
  return tokenizer(batch["verse_text"], padding=True, truncation=True)
```

```python
print(tokenize(sentiment['train'][:3]))
```

```python
sentiment_encoded = sentiment.map(tokenize, batched=True, batch_size=None)
```

```python
print(sentiment_encoded["train"].column_names)
```

```python
print(sentiment_encoded["train"])
```

```python
string_arr = [
    "Only those who will risk going too far can possibly find out how far one can go.",
    "Baby shark, doo doo doo doo doo doo, Baby shark!"
]
```

```python
inputs = tokenizer(string_arr, padding=True, truncation=True, return_tensors='pt')
```

```python
print(inputs)
```

```python
type(inputs['input_ids'])
```

```python
from transformers.tools.text_classification import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
outputs = model(**inputs)
```

```python
outputs.logits
```

```python
import torch
predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
```

```python
predictions
```

```python
model.config.id2label
```
