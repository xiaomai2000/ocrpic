```python
import dl_translate as dlt

mt = dlt.TranslationModel()  # Slow when you load it for the first time

text_hi = "संयुक्त राष्ट्र के प्रमुख का कहना है कि सीरिया में कोई सैन्य समाधान नहीं है"
mt.translate(text_hi, source=dlt.lang.HINDI, target=dlt.lang.ENGLISH)
```

```python
mt.translate(text_hi, source=dlt.lang.HINDI, target=dlt.lang.CHINESE)
```

```python
text_thai = "ขนมปัง"   #面包
print(datetime.datetime.now())
print(mt.translate(text_thai, source=dlt.lang.THAI, target=dlt.lang.CHINESE))
print(datetime.datetime.now())
print(mt.translate(text_thai, source=dlt.lang.THAI, target=dlt.lang.ENGLISH))
print(datetime.datetime.now())
```

```python
# Or NLLB-200 (faster and has 200 languages)
mt_nllb = dlt.TranslationModel("facebook/nllb-200-distilled-600M")
```

```python
#mt.available_codes()
#https://blog.csdn.net/u010302327/article/details/131183852
# tha_Thai
# zho_Hans  - Chinease simplified
# zho_Hant  - Chinese traditional
# mal_Mlym  - Malayalam
# kor_Hang  - Korean
# jpn_Jpan  - Japanese
# eng_Latn  - English
```

```python
text_thai = "ขนมปัง"   #面包
print(datetime.datetime.now())
print(mt_nllb.translate(text_thai, source='tha_Thai', target='zho_Hans'))
print(datetime.datetime.now())
print(mt_nllb.translate(text_thai, source='tha_Thai', target='eng_Latn'))
print(datetime.datetime.now())
```
