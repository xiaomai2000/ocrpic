```python
import datetime
```

```python
import dl_translate as dlt
print('Start to load model at: ', datetime.datetime.now())
mt = dlt.TranslationModel("./hfmodels/cached_model_m2m100/", model_family="m2m100")  # Slow when you load it for the first time
print('Load model done     at: ', datetime.datetime.now())
```

```python
text_hi = "संयुक्त राष्ट्र के प्रमुख का कहना है कि सीरिया में कोई सैन्य समाधान नहीं है"
mt.translate(text_hi, source=dlt.lang.HINDI, target=dlt.lang.ENGLISH)
```

```python
mt.translate(text_hi, source=dlt.lang.HINDI, target=dlt.lang.CHINESE)
```

```python
text_thai = "ขนมปัง"   #面包
print(datetime.datetime.now())
print(mt.translate(text_thai, source=dlt.lang.THAI, target=dlt.lang.CHINESE))
print(datetime.datetime.now())
print(mt.translate(text_thai, source=dlt.lang.THAI, target=dlt.lang.ENGLISH))
print(datetime.datetime.now())
```

```python
# Try batch-size:
# Laptop CPU only: 5s

text_thai = "ขนมปัง | ไอศครีม | บัตรโทรศัพท์  | น้ำแร่ | กาแฟ "   
print(datetime.datetime.now())
print(mt.translate(text_thai, source=dlt.lang.THAI, target=dlt.lang.ENGLISH, batch_size=5))
print(datetime.datetime.now())
```
